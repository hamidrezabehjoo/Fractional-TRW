```python
t = np.linspace(0,1,21)
C1 = np.loadtxt('C1.txt')
c1 = 64.09800002957432
interp_func = interp1d(C1, t)
t1 = interp_func(c1)

C2 = np.loadtxt('C2.txt')
c2 = 67.75038899271817
interp_func = interp1d(C2, t)
t2 = interp_func(c2)

C3 = np.loadtxt('C3.txt')
c3 = 72.23962614711006
interp_func = interp1d(C3, t)
t3 = interp_func(c3)

C4 = np.loadtxt('C4.txt')
c4 = 63.073683782911786
interp_func = interp1d(C4, t)
t4 = interp_func(c4)

C5 = np.loadtxt('C5.txt')
c5 = 70.96794336122501
interp_func = interp1d(C5, t)
t5 = interp_func(c5)

fig, ax1 = plt.subplots(figsize=(6, 4))
ax1.plot(t,C1,'g', t1, c1, 'g.', markersize=14, markerfacecolor='white', linestyle='solid', lw='2')
ax1.plot(t,C2,'r', t2, c2, 'r.', markersize=14, markerfacecolor='white', linestyle='solid', lw='2')
ax1.plot(t,C3,'b', t3, c3, 'b.', markersize=14, markerfacecolor='white', linestyle='solid', lw='2')
ax1.plot(t,C4,'k', t4, c4, 'k.', markersize=14, markerfacecolor='white', linestyle='solid', lw='2')
ax1.plot(t,C5,'m', t5, c5, 'm.', markersize=14, markerfacecolor='white', linestyle='solid', lw='2')

ax1.set_ylabel("$\log Z^{(\lambda)}$", fontsize=14)
ax1.set_xlabel("$\lambda$", fontsize=14)
ax1.set_xlim([0, 1])
ax1.set_ylim([62, 73])
plt.savefig('complete_graph.pdf',format="pdf", bbox_inches="tight")
```
