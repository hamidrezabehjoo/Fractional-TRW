```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def banana(X, b=0.03):
    """Twist the second column of X into a banana."""
    X = [x for x in X.copy()]
    X[1] += b * X[0]**2 - 100 * b
    return X

C = [
    [50,1],
    [1,1]
]
bananasamples = banana(np.random.multivariate_normal([0, 0], C, size=500).T)
```

```python
#plt.plot( bananasamples[0], bananasamples[1],'.' )
plt.scatter( bananasamples[0], bananasamples[1], alpha=0.5)
plt.show()
```
