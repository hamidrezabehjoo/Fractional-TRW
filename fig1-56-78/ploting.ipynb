```python
t = np.linspace(0,1,21)
grad = np.loadtxt('G.txt')
Z = np.loadtxt('Z.txt')
C = np.loadtxt('C.txt')
```

```python
from matplotlib.ticker import FormatStrFormatter
```

```python
COLOR1 = 'k'
COLOR2 = 'b'


plt.rc('xtick', labelsize=24) 
plt.rc('ytick', labelsize=24) 

fig, ax1 = plt.subplots(figsize=(8, 6))
ax2 = ax1.twinx()

ax1.plot(t,Z,'-o', color= COLOR1, lw =4, label='first')
ax2.plot(t,C,'-o', color= COLOR2, lw =4, label='second')
#ax2.plot(t,np.zeros(21), '--' )

ax1.set_ylabel("$\log Z^{(\lambda)}$",color=COLOR1, fontsize=26)
ax1.tick_params(axis="y", labelcolor=COLOR1)

#ax2.set_ylabel("$\log \tilde{Z}^{(\lambda)}$", color=COLOR2, fontsize=26)
ax2.set_ylabel("$\\log \\tilde{Z}^{(\\lambda)}$", color=COLOR2, fontsize=26)
ax2.tick_params(axis="y", labelcolor=COLOR2)

ax1.set_xlabel("$\lambda$", fontsize=26)
plt.ticklabel_format(axis='both', style='sci')

plt.tight_layout()
#plt.grid()

plt.savefig('a.pdf')
```

```python
h = t[1] - t[0]
gg = np.zeros(21)
for i in range(1,20):
    gg[i] = (grad[i+1] - grad[i-1]) / 2*h
```

```python
COLOR1 = 'k'
COLOR2 = 'b'

plt.rc('xtick', labelsize=24) 
plt.rc('ytick', labelsize=24) 

fig, ax1 = plt.subplots(figsize=(8, 6))
ax2 = ax1.twinx()


ax1.plot(t,grad,'-o', lw=4, color= COLOR1)
ax2.plot(t[1:20],gg[1:20],'-o', lw =4, color= COLOR2)
#ax2.plot(t,np.zeros(21), '--' )

ax1.set_ylabel("$dF^{(\lambda)}/d\lambda$",color=COLOR1, fontsize=26)
ax1.tick_params(axis="y", labelcolor=COLOR1)

ax2.set_ylabel("$d^2F^{(\lambda)}/d\lambda^2$", color=COLOR2, fontsize=26)
ax2.tick_params(axis="y", labelcolor=COLOR2)

ax1.set_xlabel("$\lambda$", fontsize=26)
#plt.xlim([0,1])
plt.tight_layout()
#plt.grid()

plt.savefig('a1.pdf')
```
