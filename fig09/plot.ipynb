```python
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
path = os.getcwd()
```

```python
path = path + '/out'
os.chdir(path)
#os.getcwd()

files = os.listdir(path)
for file in files:
    print(file)
```

```python
t = np.linspace(0,1,21)
C1 = np.loadtxt('C1.txt')
c1 = 68.88364413800622
interp_func = interp1d(C1, t)
t1 = interp_func(c1)

C2 = np.loadtxt('C2.txt')
c2 = 68.77363682009717
interp_func = interp1d(C2, t)
t2 = interp_func(c2)

C3 = np.loadtxt('C3.txt')
c3 = 65.68770247030346
interp_func = interp1d(C3, t)
t3 = interp_func(c3)

C4 = np.loadtxt('C4.txt')
c4 = 73.2999098189046
interp_func = interp1d(C4, t)
t4 = interp_func(c4)

C5 = np.loadtxt('C5.txt')
c5 = 66.4679140724479
interp_func = interp1d(C5, t)
t5 = interp_func(c5)

fig, ax1 = plt.subplots(figsize=(8, 6))
plt.rc('xtick', labelsize=24) 
plt.rc('ytick', labelsize=24) 


ax1.plot(t,C1,'g', t1, c1, 'g.', markersize=20, markerfacecolor='white', linestyle='solid', lw='4')
ax1.plot(t,C2,'r', t2, c2, 'r.', markersize=20, markerfacecolor='white', linestyle='solid', lw='4')
ax1.plot(t,C3,'b', t3, c3, 'b.', markersize=20, markerfacecolor='white', linestyle='solid', lw='4')
ax1.plot(t,C4,'k', t4, c4, 'k.', markersize=20, markerfacecolor='white', linestyle='solid', lw='4')
ax1.plot(t,C5,'m', t5, c5, 'm.', markersize=20, markerfacecolor='white', linestyle='solid', lw='4')

ax1.set_ylabel("$\log Z^{(\lambda)}$", fontsize=26)
ax1.set_xlabel("$\lambda$", fontsize=26)
ax1.set_xlim([0, 1])
ax1.set_ylim([64, 70])
plt.savefig('complete_graph.pdf',format="pdf", bbox_inches="tight")
```
