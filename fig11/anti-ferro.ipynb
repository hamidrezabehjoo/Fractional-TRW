```python
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
path = os.getcwd()
print(path)
```

```python
path1 = path + '/out'


files = os.listdir(path1)
for file in files:
    print(file)
```

```python
t = np.linspace(0,1,21)
Z1 = np.loadtxt('Z1.txt')
z1 =  44.725374888351205

Z2 = np.loadtxt('Z2.txt')
z2 =  45.272550867731026



plt.rc('xtick', labelsize=24) 
plt.rc('ytick', labelsize=24) 

fig, ax1 = plt.subplots(figsize=(8, 6))
ax1.plot(t,Z1,'ok', linestyle='solid', lw=4)
ax1.plot(t, z1*np.ones(21) ,'--k', lw =4)

ax1.plot(t,Z2,'ob', linestyle='solid', lw=4)
ax1.plot(t, z2*np.ones(21) ,'--b', lw =4 )

ax1.set_ylabel("$\log Z^{(\lambda)}$", fontsize=26)
ax1.set_xlabel("$\lambda$", fontsize=26)
ax1.set_xlim([0, 1])

plt.ticklabel_format(axis='both', style='sci')
plt.tight_layout()
plt.savefig('Z-anti.pdf')
plt.show()
```
