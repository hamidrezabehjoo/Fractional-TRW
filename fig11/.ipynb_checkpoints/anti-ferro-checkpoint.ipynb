```python
t = np.linspace(0.05,1,21)
Z1 = np.loadtxt('Z1.txt')
z1 =  42.55328625887392

Z2 = np.loadtxt('Z2.txt')
z2 =  42.79544961154541


fig, ax1 = plt.subplots(figsize=(6, 4))
ax1.plot(t,Z1,'ok', linestyle='solid', lw=2)
ax1.plot(t, z1*np.ones(21) ,'--k', lw =2)

ax1.plot(t,Z2,'ob', linestyle='solid', lw=2)
ax1.plot(t, z2*np.ones(21) ,'--b', lw =2 )

ax1.set_ylabel("$\log Z^{(\lambda)}$", fontsize=14)
ax1.set_xlabel("$\lambda$", fontsize=14)
ax1.set_xlim([0, 1])
plt.savefig('Z-anti.pdf')
```
