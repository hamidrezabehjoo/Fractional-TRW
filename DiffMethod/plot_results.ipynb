```python
# z_true:	 53.793238654852885 4x4 grid

# n = 1  52.35062555666989 1.7952373969190198
# n = 2  53.43518116865229 0.8183566850753172
# n = 3  53.75238361085947 0.38607146698260475
# n = 4  53.785056934273555 0.12065188700374


# t=0.67
# n = 1 52.39342344065423 0.8957561645772506
# n = 2 53.431203169965966 0.3340966843414782
# n = 3 53.786633716952556 0.13933432050970862
# n = 4 53.79588781579523 0.05188350566412602


import matplotlib.pyplot as plt
import numpy as np

# Data
n_values = [1, 2, 3, 4]

# t = 0 data
means_t0 = [52.35062555666989, 53.43518116865229, 53.75238361085947, 53.785056934273555]
variances_t0 = [1.7952373969190198, 0.8183566850753172, 0.38607146698260475, 0.12065188700374]

# t = 0.67 data
means_t067 = [52.39342344065423, 53.431203169965966, 53.786633716952556, 53.79588781579523]
variances_t067 = [0.8957561645772506, 0.3340966843414782, 0.13933432050970862, 0.02288350566412602]

# Calculate standard deviations from variances
std_devs_t0 = np.sqrt(variances_t0)
std_devs_t067 = np.sqrt(variances_t067)

# True Z value
z_true = [53.793238654852885] * 4

# Plot both datasets on the same figure
fig, ax = plt.subplots(figsize=(8, 6))

# Plot for t = 0
ax.errorbar(n_values, means_t0, yerr=std_devs_t0, fmt='o', capsize=5, capthick=2, 
            ecolor='red', color='red', markerfacecolor='red', markeredgecolor='black', 
            markersize=9, linewidth=4, label=r'$\lambda = 0$')

# Plot for λ = λ*
ax.errorbar(n_values, means_t067, yerr=std_devs_t067, fmt='s', capsize=5, capthick=2, 
            ecolor='blue', color='blue', markerfacecolor='blue', markeredgecolor='black', 
            markersize=9, linewidth=4, label=r'$\lambda = \lambda_*$')

# Plot the True Z line
ax.plot(n_values, z_true, '-.', color='black', linewidth=4, label='True Z')

# Customize the plot
ax.set_xlabel('i', fontsize=26)
ax.set_ylabel(r'$\log Z$', fontsize=26)
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.legend(fontsize=12, loc='lower right')

# Set x-axis to show integer values
ax.set_xticks(n_values)
ax.set_xticklabels(n_values, fontsize=24)

# Set y-axis tick labels
y_ticks = np.arange(50, 55, 1.0)
ax.set_yticks(y_ticks)
ax.set_yticklabels([f'{y:.1f}' for y in y_ticks], fontsize=24)

# Adjust y-axis to show full error bars
#plt.ylim(50, 55)

# Add gridlines for better readability
ax.yaxis.grid(True, linestyle='--', which='major', color='grey', alpha=0.5)


# Show the plot
plt.tight_layout()
plt.savefig('44.pdf')
plt.show()
```
