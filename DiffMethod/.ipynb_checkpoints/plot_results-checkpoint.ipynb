```python
import matplotlib.pyplot as plt
import numpy as np

# Data
n_values = [1, 2, 3, 4]

# t = 0 data
means_t0 = [93.21776986633327, 94.72176480350386, 95.31867171756811, 95.47817100395946]
variances_t0 = [1.717955004060111, 0.8243213796519614, 0.4566745259942479, 0.1566745259942479]

# t = 0.74 data
means_t074 = [93.34571253350201, 94.88505546362656, 95.47795661798312, 95.47817100395946]
variances_t074 = [1.0181164674145127, 0.30869946063257636, 0.10863892148886011, 0.008685318121079568]

z_true = [95.47817100395946, 95.47817100395946, 95.47817100395946, 95.47817100395946]

# Calculate standard deviations from variances
std_devs_t0 = np.sqrt(variances_t0)
std_devs_t074 = np.sqrt(variances_t074)

# Plot both datasets on the same figure
fig, ax = plt.subplots(figsize=(8, 6))

# Plot for t = 0
ax.errorbar(n_values, means_t0, yerr=std_devs_t0, fmt='o', capsize=5, capthick=2, 
            ecolor='red', color='red', markerfacecolor='red', markeredgecolor='black', 
            markersize=9, linewidth=4, label=r'$\lambda = 0$')

# Plot for \lambda = \lambda^*
ax.errorbar(n_values, means_t074, yerr=std_devs_t074, fmt='s', capsize=5, capthick=2, 
            ecolor='blue', color='blue', markerfacecolor='blue', markeredgecolor='black', 
            markersize=9, linewidth=4, label=r'$\lambda = \lambda^*$')

# Plot the True z line
ax.plot(n_values, z_true, '-.', color='black', linewidth=4, label='True Z')

# Customize the plot
ax.set_xlabel('i', fontsize=26)
ax.set_ylabel(r'$\log Z$', fontsize=26)
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.legend(fontsize=12, loc='center right', bbox_to_anchor=(1, 0.5))

# Set x-axis to show integer values
ax.set_xticks(n_values)
ax.set_xticklabels(n_values, fontsize=24)
ax.set_yticklabels(np.arange(91.5, 97, 0.5), fontsize=24)

# Adjust y-axis to show full error bars
plt.ylim(91.5, 96.5)
#plt.xlim(0.5, 4.5)

# Add gridlines for better readability
ax.yaxis.grid(True, linestyle='--', which='major', color='grey', alpha=0.5)

# Show the plot
plt.tight_layout()
plt.savefig('diffMethod.pdf')

plt.show()
```
