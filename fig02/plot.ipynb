```python
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
path = os.getcwd()
print(path)
```

```python
path1 = path + '/out'

files = os.listdir(path1)
for file in files:
    print(file)
```

```python
path = path1 + '/10'
os.chdir(path)

t = np.linspace(0,1,21)

Z1 = np.loadtxt('Z1.txt')
f1 = interp1d(Z1, t)
z1 = 4.2022
t1 = f1(z1)

Z2 = np.loadtxt('Z2.txt')
f2 = interp1d(Z2, t)
z2 = 4.37
t2 = f2(z2)

Z3 = np.loadtxt('Z3.txt')
f3 = interp1d(Z3, t)
z3 = 4.28
t3 = f3(z3)

Z4 = np.loadtxt('Z4.txt')
f4 = interp1d(Z4, t)
z4 = 4.222
t4 = f4(z4)



plt.rc('xtick', labelsize=24) 
plt.rc('ytick', labelsize=24) 
#plt.figure(figsize=(8,6))

fig, ax1 = plt.subplots(figsize=(8, 6))
plt.ticklabel_format(axis='both', style='sci', scilimits=(0,0))

ax1.plot(t,Z1,'og', linestyle='solid', lw='4')
ax1.plot(t1,z1,'og', markersize=14, markerfacecolor='white')

ax1.plot(t,Z2,'sr', linestyle='solid', lw='4')
ax1.plot(t2,z2,'sr', markersize=14, markerfacecolor='white')

ax1.plot(t,Z3,'vb', linestyle='solid', lw='4')
ax1.plot(t3,z3,'vb', markersize=14, markerfacecolor='white')

ax1.plot(t,Z4,'dk', linestyle='solid', lw='4')
ax1.plot(t4,z4,'dk', markersize=14, markerfacecolor='white')

ax1.set_ylabel(r"$\frac{\log Z^{(\lambda)}}{N}$", fontsize=32)
ax1.set_xlabel("$\lambda$", fontsize=26)
ax1.set_xlim([0, 1])
#ax1.set_ylim([4.1, 4.61])

plt.ticklabel_format(axis='both', style='sci')
plt.tight_layout()
plt.savefig('planar_graph.pdf')
```
