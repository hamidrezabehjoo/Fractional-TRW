```python
path = '/home/hamidreza/Code/fig02/out/10'
os.chdir(path)

t = np.linspace(0,1,21)

Z1 = np.loadtxt('Z1.txt')
f1 = interp1d(Z1, t)
z1 = 4.398275031576794731e+02
t1 = f1(z1)

Z2 = np.loadtxt('Z2.txt')
f2 = interp1d(Z2, t)
z2 = 4.112955629392240553e+02
t2 = f2(z2)

Z3 = np.loadtxt('Z3.txt')
f3 = interp1d(Z3, t)
z3 = 4.127963797687109491e+02
t3 = f3(z3)

Z4 = np.loadtxt('Z4.txt')
f4 = interp1d(Z4, t)
z4 = 4.127137745990274880e+02
t4 = f4(z4)




fig, ax1 = plt.subplots(figsize=(6, 4))
ax1.plot(t,Z1,'og', linestyle='solid', lw='2')
ax1.plot(t1,z1,'og', markersize=10, markerfacecolor='white')

ax1.plot(t,Z2,'sr', linestyle='solid', lw='2')
ax1.plot(t2,z2,'sr', markersize=10, markerfacecolor='white')

ax1.plot(t,Z3,'vb', linestyle='solid', lw='2')
ax1.plot(t3,z3,'vb', markersize=10, markerfacecolor='white')

ax1.plot(t,Z4,'dk', linestyle='solid', lw='2')
ax1.plot(t4,z4,'dk', markersize=10, markerfacecolor='white')

ax1.set_ylabel("$\log Z^{(\lambda)}$", fontsize=14)
ax1.set_xlabel("$\lambda$", fontsize=14)
ax1.set_xlim([0, 1])
plt.savefig('planar_graph.pdf')
```
