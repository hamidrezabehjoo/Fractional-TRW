```python
t = np.linspace(0,1,21)
grad = np.loadtxt('G.txt')
Z = np.loadtxt('Z.txt')
C = np.loadtxt('C.txt')
```

```python
COLOR1 = 'g'
COLOR2 = 'b'

fig, ax1 = plt.subplots(figsize=(7, 4))
ax2 = ax1.twinx()


ax1.plot(t,Z,'o', color= COLOR1, label='first')
ax2.plot(t,C,'o', color= COLOR2, label='second')
#ax2.plot(t,np.zeros(21), '--' )

ax1.set_ylabel("$\log Z^{(\lambda)}$",color=COLOR1, fontsize=14)
ax1.tick_params(axis="y", labelcolor=COLOR1)

ax2.set_ylabel("$\log \mathcal{Z}^{(\lambda)}$", color=COLOR2, fontsize=14)
ax2.tick_params(axis="y", labelcolor=COLOR2)

ax1.set_xlabel("$\lambda$")
#plt.grid()

plt.savefig('a.pdf')
```

```python
h = t[1] - t[0]
gg = np.zeros(21)
for i in range(1,20):
    gg[i] = (grad[i+1] - grad[i-1]) / 2*h
```

```python
COLOR1 = 'g'
COLOR2 = 'b'

fig, ax1 = plt.subplots(figsize=(7, 4))
ax2 = ax1.twinx()


ax1.plot(t,grad,'o', color= COLOR1)
ax2.plot(t[1:20],gg[1:20],'o', color= COLOR2)
#ax2.plot(t,np.zeros(21), '--' )

ax1.set_ylabel("$dF^{(\lambda)}/d\lambda$",color=COLOR1, fontsize=14)
ax1.tick_params(axis="y", labelcolor=COLOR1)

ax2.set_ylabel("$d^2F^{(\lambda)}/d\lambda^2$", color=COLOR2, fontsize=14)
ax2.tick_params(axis="y", labelcolor=COLOR2)

ax1.set_xlabel("$\lambda$")
#plt.grid()

plt.savefig('a1.pdf')
```
